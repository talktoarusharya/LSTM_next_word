```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv('/content/qoute_dataset.csv')
```

```python
df.head()
```

```python
sentences = df['quote'].head(100);
sentences.head()
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(sentences)
```

```python
tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)
```

```python
input_sequences = []
for sentence in sentences:
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1, len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])


```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding = 'pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
len(tokenizer.index_word)
```

```python
y = to_categorical(y,num_classes=718)
```

```python
y.shape
```

```python
y
```

```python
model = Sequential()
model.add(Embedding(718, 100,input_length=198,input_shape=(198,)))
model.add(LSTM(150))
model.add(Dense(718, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100,batch_size=128)
```

```python
text = "The real lover"


reverse_word_index = {v: k for k, v in tokenizer.word_index.items()}

for i in range(20):
    token_text = tokenizer.texts_to_sequences([text])[0]
    padded_token_text = pad_sequences([token_text], maxlen=198, padding='pre')

    preds = model.predict(padded_token_text, verbose=0)[0]
    for word in text.lower().split():
        if word in tokenizer.word_index:
            idx = tokenizer.word_index[word]
            preds[idx] = preds[idx] * 0.01

    pos = np.argmax(preds)

    word = reverse_word_index.get(pos, "")
    text = text + " " + word
    print(text)
```
